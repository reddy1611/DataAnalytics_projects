Question wise toppers

```python
toppers = {}
for col in ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']:
    toppers[col] = df[df[col] == df[col].max()].index.tolist()

print("Question-wise Toppers:", toppers)
```

AVERAGE SCORE FOR EACH QUESTION

```python
average_scores = df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']].mean()
average_scores.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Scores per Question")
plt.xlabel("Questions")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()
```

TOTAL SCORE DISTRIBUTION

```python
df['Total'].plot(kind='hist', bins=10, color='orange', edgecolor='black', alpha=0.7)
plt.title("Distribution of Total Scores")
plt.xlabel("Total Scores")
plt.ylabel("Frequency")
plt.show()
```

BOXPLOT SHOWS

```python
sns.boxplot(data=df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']])
plt.title("Score Spread per Question")
plt.xlabel("Questions")
plt.ylabel("Scores")
plt.show()
```

PIC CHART SHOWING THAT STUDENTS SCORING FULL MARKS

```python
full_marks = df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']].apply(lambda row: any(row == 10), axis=1).sum()
no_full_marks = len(df) - full_marks

data = [full_marks, no_full_marks]
labels = ['Full Marks in Any Question', 'No Full Marks']

plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, colors=['GREEN', 'YELLOW'])
plt.title("Proportion of Students Scoring Full Marks")
plt.show()
```

PIC CHART SHOWING AVERAGE SCORES BYE QUESTION

```python
average_scores = df.iloc[:, 1:].mean()
plt.figure(figsize=(8, 8))
plt.pie(average_scores, labels=average_scores.index, autopct="%1.1f%%", startangle=140)
plt.title("Question-Wise Average Scores")
plt.show()
```
