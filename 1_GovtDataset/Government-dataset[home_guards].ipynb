## Dataset Description

The dataset outlines the home guard strength across various districts for April 2024, detailing the number of units, allotted personnel, and actual  deployed strength. It provides a breakdown of male and female home guards, helping to analyze gender distribution in the workforce. Additionally, the dataset highlights vacancies. This data can be useful for assessing staffing efficiency, identifying shortages, and making informed decisions regarding deployment and recruitment.

```python
plt.figure(figsize=(12, 6))
plt.bar(df['NAME OF THE DISTRICT'], df['ACTUAL STRENGTH'], label="Actual Strength", alpha=0.7)
plt.bar(df['NAME OF THE DISTRICT'], df['ALLOTTED STRENGTH'], label="Allotted Strength", alpha=0.5)
plt.xticks(rotation=90)
plt.xlabel("District")
plt.ylabel("Strength")
plt.title("Actual vs. Allotted Home Guard Strength")
plt.legend()
plt.show()
```
